```python
import pandas as pd
import numpy as np

df = pd.read_csv("customer_churn_dirty (1).csv")
```

```python
columns = ['Customer_ID', 'Tenure_Months', 'Age', 'Gender', 'Contract_Type', 'Monthly_Charges', 'Total_Charges',
           'Payment_Method', 'Internet_Service', 'Has_Multiple_Lines', 'Streaming_TV', 'Churn', 'Reason_for_Churn']
df = pd.DataFrame(df, columns=columns)
```

```python
churn_df = df.drop(columns=['Payment_Method', 'Customer_ID'])
```

```python
churn_df['Gender'] = churn_df['Gender'].str.strip().str.lower()

churn_df['Gender'] = churn_df['Gender'].replace(['male', 'm'], 1)
# churn_df['Gender'] = churn_df['Gender'].replace('M', 1)
churn_df['Gender'] = churn_df['Gender'].replace('female', 0)
```

```python
target1 = ['monthly', 'm', 'monthy']
target2 = ['yearly', 'y']
target3 = ['two-year', '2-year', 'two year']
churn_df['Contract_Type'] = churn_df['Contract_Type'].str.strip().str.lower()

churn_df['Contract_Type'] = churn_df['Contract_Type'].replace(target1, 1)
churn_df['Contract_Type'] = churn_df['Contract_Type'].replace(target2, 2)
churn_df['Contract_Type'] = churn_df['Contract_Type'].replace(target3, 3)
```

```python
churn_df['Tenure_Months'] = churn_df['Tenure_Months'].abs()
```
